```python
# 1. "x"와 "y"열을 사용하여 산점도를 그리세요.
# 2. "category"열을 사용하여 데이터 포인트를 색으로 구분하세요.
df4 = pd.DataFrame({"x" : [1,2,3,4,5], "y":[10,20,15,25,30], "category":["A","B","A","B","A"]})
sns.scatterplot(x="x",y="y",hue="category", data=df4)
```

## **실습문제**

주어진 데이터프레임에서 "category"열은 데이터 포인트의 범주를, "value"열은 해당 데이터 포인트이 값을 나타냅니다. 주어진 데이터를 이용하여 seaborn의 countplot()함수를 사용하여 다음을 수행하세요.

```python
# 1. "category"열을 사용하여 각 카테고리별로 데이터의 개수를 세어서 막대 그래프로 시각화하세요.
data = pd.DataFrame({"category" : ["A","B","A","C","B","A","A","C","B","B"], "value" : [10,20,15,25,30,12,18,22,28,35]})
sns.countplot(x="category", hue="category", data=data)
plt.show()
```
