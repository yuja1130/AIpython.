```python
# 1. 데이터를 이용하여 데이터프레임 생성
data = {"학생" : ["John", "Alice", "Bob", "Emily"], "수학" : [90, 87, 78, 92], "영어" : [85, 91, 80, 88], "과학" : [88,89,85,82]}
df = pd.DataFrame(data)
df
```

```python
# 2. 생성한 데이터프레임의 요약 통계 정보를 출력하세요
df.describe()
```

```python
# 3. 생성한 데이터프레임에서 "수학" 점수가 가장 높은 학생의 정보를 출력하세요.(max() 메서드 활용)
df[df["수학"] == df["수학"].max()]
```

```python
# 4. 생성한 데이터프레임에서 "수학" 점수가 90점 이상인 학생의 정보를 출력하세요.(조건연산자 활용)
df[df["수학"] >= 90]
```
